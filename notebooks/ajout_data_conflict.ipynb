```python
# Périodes cibles : 1990, 1995, 2000, 2005, 2010
periodes = [1990, 1995, 2000, 2005, 2010,2015]

# Périodes cibles : 1990, 1995, 2000, 2005, 2010
periodes = [1990, 1995, 2000, 2005, 2010,2015]

# Computation : conflit pour ce pays cette année ?
conflict_annual = df_ucdp.groupby(['year', 'iso3']).agg(
    conflict_dummy = ('conflict_id', 'nunique'),
    intensity_max  = ('intensity_level', 'max')
).reset_index()

conflict_annual['conflict_dummy'] = (conflict_annual['conflict_dummy'] > 0).astype(int)

# Pour chaque pays et chaque période t,
# conflict_5y = 1 si conflit dans [t-5, t-1]
# N.B Pour l'indicateur intensityé : on l'intensité maximale observée sur toute la fenêtre de 5 ans :
rows = []
for iso in conflict_annual['iso3'].unique():
    pays = conflict_annual[conflict_annual['iso3'] == iso].set_index('year')
    for t in periodes:
        fenetre = pays.loc[pays.index.isin(range(t-5, t))]
        
        conflict_5y   = int(fenetre['conflict_dummy'].sum() > 0)
        intensity_5y  = int(fenetre['intensity_max'].max()) if conflict_5y else 0
        
        rows.append({
            'iso3'        : iso,
            'year'        : t,
            'conflict_5y' : conflict_5y,
            'intensity_5y': intensity_5y
        })

indicators = pd.DataFrame(rows)
print(indicators.head(20))
```

```python
# Merge destination (j)
gravity = gravity.merge(
    indicators.rename(columns={
        'conflict_5y' : 'conflict_5y_j',
        'intensity_5y': 'intensity_5y_j'
    }),
    left_on=['iso3_d', 'year'],
    right_on=['iso3', 'year'],
    how='left'
).drop(columns=['iso3'])
```

```python
# NaN → 0
gravity[['conflict_5y_i', 'intensity_5y_i',
         'conflict_5y_j', 'intensity_5y_j']] = \
gravity[['conflict_5y_i', 'intensity_5y_i',
         'conflict_5y_j', 'intensity_5y_j']].fillna(0).astype(int)
```
